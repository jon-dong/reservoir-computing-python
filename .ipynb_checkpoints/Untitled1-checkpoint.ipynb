```python
from oct2py import octave
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt

N = 1000;  L = 22.;  h = 0.25;  nstp = 1000;
a0 = np.zeros([N-2,1]);  a0[0:4] = 0.6; # just some initial condition
# fig1 = figure('pos',[5 550 600 400],'color','w'); plot(tt,at,'.-');
[tt, at] = octave.feval('ksfmstp', a0, L, h, nstp, 1, nout=2)
[x, ut] = octave.feval('ksfm2real', at, L, nout=2);
# fig2 = figure('pos',[5 270 600 200],'color','w');
# pcolor(tt,x,ut); shading interp; caxis([-3 3]);
# title('Solution u(x,t) of Kuramoto-Sivashinsky equation, system size L = 22:');
# xlabel('Time'); ylabel('x','rotat',0);
im=plt.pcolormesh(ut, cmap='gist_rainbow')
plt.colorbar(im)
plt.show
```
