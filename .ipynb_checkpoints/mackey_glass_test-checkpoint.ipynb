```python
#!/usr/bin/python

from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

# import os
# import sys
import numpy as np

import time

from reservoir import Reservoir
import data

%load_ext autoreload
%autoreload 1
%aimport data
%aimport reservoir

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
b = Reservoir(n_res=1000, input_scale=2, train_method='ridge', train_param=5,
              weights_type='complex gaussian', random_projection='simulation',
              activation_fun='phase', activation_param=1, encoding_method='phase',
              input_dim=1, forget=0)
```

```python
roll_num = 500
input_data = data.mackey_glass(sequence_length=4000)
y = data.roll_and_concat(input_data, roll_num=roll_num)
input_data = input_data[:,:-roll_num,:]
y = y[:,:-roll_num,:]
b.fit(input_data, y)
```

```python
input_data = data.mackey_glass(sequence_length=1000)
y = data.roll_and_concat(input_data, roll_num=roll_num)[:, b.forget:, :]
pred_output, valid_score = b.score(input_data, y)
```

```python
# n_input = round(pred_output.shape[0]/2)-100
n_input = 10
all_data = np.concatenate((y[0, 0:n_input, 0], pred_output[n_input, :n_input+roll_num]))
plt.figure(1)
plt.plot(all_data)
plt.plot(y[0, 0:n_input+roll_num, 0])
plt.show()
plt.figure(2)
plt.plot(y[0, 0:n_input+roll_num, 0]-all_data)
# plt.plot(np.linspace(n_input,n_input,100),np.linspace(-0.6,0.6,100))
plt.show()
```

```python
fig, axs = plt.subplots(y.shape[-1], 1, sharex=True, figsize=(10,10))
fig.subplots_adjust(hspace=0)
for i in range(y.shape[-1]):
    axs[i].plot(y[0,:,i])
    axs[i].plot(pred_output[:,i])
plt.show()
```

```python
# rms for binary encoding

length_input = pred_output.shape[0]-roll_num
rms = np.zeros((length_input, roll_num))
for n_input in range(1, length_input):
    for n_pred in range(1, roll_num):
        d1 = pred_output[n_input, :n_pred]
        d2 = y[0, n_input:n_input+n_pred, 0]
        rms[n_input, n_pred] = np.sqrt(1./(n_pred)*np.sum((d1 - d2)**2))

plt.figure(1,figsize=(10,2))
plt.title('rms for binary encoding (10 bit)')
plt.ylabel('input length')
plt.xlabel('Prediction length')
im=plt.pcolormesh(rms[:,:], cmap='gist_heat')
plt.colorbar(im)
plt.show;

plt.figure(2,figsize=(8.1,2))
plt.ylabel('average rms by input length')
plt.xlabel('Prediction length')
plt.plot(np.sum(rms[:,:], axis=0)/length_input)
plt.show;
```

```python
# rms for phase encoding

length_input = pred_output.shape[0]-roll_num
rms = np.zeros((length_input, roll_num))
for n_input in range(1, length_input):
    for n_pred in range(1, roll_num):
        d1 = pred_output[n_input, :n_pred]
        d2 = y[0, n_input:n_input+n_pred, 0]
        rms[n_input, n_pred] = np.sqrt(1./(n_pred)*np.sum((d1 - d2)**2))

plt.figure(1,figsize=(10,2))
plt.title('rms for phase encoding (8 bit)')
plt.ylabel('input length')
plt.xlabel('Prediction length')
im=plt.pcolormesh(rms[1:,1:], cmap='gist_heat')
plt.colorbar(im)
plt.show;

plt.figure(2,figsize=(8.1,2))
plt.ylabel('average rms by input length')
plt.xlabel('Prediction length')
plt.plot(np.sum(rms[1:,1:], axis=0)/length_input)
plt.show;
```

```python
plt.figure(2,figsize=(8.1,2))
plt.ylabel('average rms by input length')
plt.xlabel('Prediction length')
plt.plot(np.sum(rms[1:,:], axis=1)/roll_num)
plt.show;
```
